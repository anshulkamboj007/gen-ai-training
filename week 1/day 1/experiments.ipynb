```python

def linear_regression(X, y):
    X = np.array(X).flatten()
    y = np.array(y).flatten()

    X_mean = np.mean(X)
    y_mean = np.mean(y)

    b1_num = 0
    b1_denom = 0
    for i in range(len(X)):
        b1_num += (X[i] - X_mean) * (y[i] - y_mean)
        b1_denom += (X[i] - X_mean) ** 2

    b1 = b1_num / b1_denom
    b0 = y_mean - b1 * X_mean
    return b0, b1

b0,b1=linear_regression(X,y)
print(b0,b1)
y_pred=b0+b1*X

```
